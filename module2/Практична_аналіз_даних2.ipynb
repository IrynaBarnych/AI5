# Завдання 3
Чи потрібно застосувати квантильний алгоритм для видалення викидів? Якщо так то застосуйте, інакше, що на вашу думку потрібно зробити?

```python
def remove_outliers(df, column_name):
  column = df[column_name]

  q1 = column.quantile(q=0.25)
  q3 = column.quantile(q=0.75)

  iqr = q3 - q1

  lower = q1 - 1.5*iqr
  upper = q3 + 1.5*iqr

  mask = (column >= lower) & (column <= upper)

  return df[mask]
```

```python
for column_name in df.columns:
  if df.dtypes[column_name] == 'object':
    continue

  df = remove_outliers(df, column_name)
```

```python
df['Temperature (C)'].plot(kind='box', vert=False, figsize=(14,6))
```

# Завдання 6
Виведіть групові ящики з вусами по одній(або декількох) категоріальних ознак. Що по них можна сказати?

```python
num_columns = df.select_dtypes(include="number").columns
for column in num_columns:
    df[[column, "Temperature (C)"]].boxplot()
    plt.title(f'Boxplot for {column} by Temperature (C)')
    plt.show()
```
